```python
from google.colab import files
uploaded = files.upload()

import zipfile, os

with zipfile.ZipFile("homer_bart.zip", 'r') as zip_ref:   # change name if needed
    zip_ref.extractall("/content/homer_bart_extracted")

print(os.listdir("/content/homer_bart_extracted"))
```

```python

import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.1,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_data = datagen.flow_from_directory(
    "/content/homer_bart_extracted",
    target_size=(64,64),
    batch_size=32,
    class_mode='binary',
    subset='training',
    seed=42,
    shuffle=True
)

test_data = datagen.flow_from_directory(
    "/content/homer_bart_extracted",
    target_size=(64,64),
    batch_size=32,
    class_mode='binary',
    subset='validation',
    seed=42,
    shuffle=False
)

model = models.Sequential([
    layers.Input(shape=(64,64,3)),
    layers.Flatten(),

    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.003)),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.003)),
    layers.BatchNormalization(),
    layers.Dropout(0.4),

    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.003)),
    layers.Dropout(0.3),

    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(train_data, epochs=100, validation_data=test_data)

test_loss, test_accuracy = model.evaluate(test_data)
print("\nğŸ”¥ FINAL TEST ACCURACY:", test_accuracy)
```
